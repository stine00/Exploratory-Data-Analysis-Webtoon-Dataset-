```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv(r'C:/Users/pc/Desktop/olive.csv')

# Intro for Part 2
print("\nLOADING AND CLEANSING")

# Display the first 5 rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(df.head(5))

# Print the column names of the DataFrame
print("\nColumn names of the DataFrame:")
print(df.columns)

# Rename the first column to 'areastring'
df.rename(columns={df.columns[0]: 'areastring'}, inplace=True)

# Print the column names after renaming
print("\nColumn names after renaming:")
print(df.columns)

# Print the regions and areas in the dataset
print("\nWhich unique regions and areas are contained in the dataset?")
print('regions\t', df['region'].unique())
print('areas\t', df['area'].unique())

# Create a crosstabulation or contingency table of the factors
print("\nCreate a crosstabulation or contingency table of the factors")
cross_tab = pd.crosstab(df.area, df.region)
print(cross_tab)

# Clean the dataset
df.head()

# Get rid of the junk numbering in df.areastring using map transformation
print("\nGet rid of the junk numbering in df.areastring using map transformation")
df.areastring = df.areastring.map(lambda x: x.split('.')[-1])
print(df.head())

# Use list indexing to access a specific subset of columns of a dataframe
print("\nUse list indexing to access a specific subset of columns of a dataframe")
subset = df[['palmitic', 'oleic']]
print(subset.head())

# To access the series of entries of a single column
print("\nTo access the series of entries of a single column")
series = df['palmitic']
print(series)

# Second example
print("\nNotice the difference in the syntax. In the first example where we used list indexing, we got a new DataFrame. In the second example, we got a series corresponding to the column.")
print("Type of df[['palmitic']]:\t", type(df[['palmitic']]))
print("Type of df['palmitic']:\t", type(df['palmitic']))

# To access the series of values of a single column
series_values = df.palmitic
print(series_values)


import matplotlib.pyplot as plt

# Intro for Part 3
print("\nQUICK INTRO TO MATPLOTLIB")

# Plots
fig = plt.figure()
plt.scatter(df.palmitic, df.linolenic)
axis = fig.gca()  # Get current axis
axis.set_title('linolenic vs palmitic')
axis.set_xlabel('palmitic')
axis.set_ylabel('linolenic')

# fig can be accessed with fig.gcf()

plt.hist(df.palmitic)

# Display the plots
plt.show()

# A more object-oriented approach using the subplots function
fig, axes = plt.subplots(figsize = (10,10), nrows = 2, ncols = 2)
axes[0][0].plot(df.palmitic, df.linolenic)
axes[0][1].plot(df.palmitic, df.linolenic, '.')
axes[1][0].scatter(df.palmitic, df.linolenic)
axes[1][1].hist(df.palmitic)
fig.tight_layout()


# Intro for Part 4
print("\n PANDAS DATA MUNGING")

#Pandas grouby
region_groupby = df.groupby('region')
print (type(region_groupby))
region_groupby.head()
```
